```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def visualize_fraction_multiplication(frac1_num, frac1_den, frac2_num, frac2_den):
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # 1. First fraction
    plt.subplot(131)
    ax1 = plt.gca()
    # Draw full square
    ax1.add_patch(Rectangle((0, 0), 1, 1, fill=False, color='black'))
    # Draw fraction part
    ax1.add_patch(Rectangle((0, 0), 1, frac1_num/frac1_den, 
                          color='blue', alpha=0.3))
    
    # Add grid lines for denominator
    for i in range(1, frac1_den):
        plt.axhline(y=i/frac1_den, color='gray', linestyle='--', alpha=0.3)
    
    plt.title(f'{frac1_num}/{frac1_den}')
    plt.xlim(-0.1, 1.1)
    plt.ylim(-0.1, 1.1)
    
    # 2. Second fraction
    plt.subplot(132)
    ax2 = plt.gca()
    # Draw full square
    ax2.add_patch(Rectangle((0, 0), 1, 1, fill=False, color='black'))
    # Draw fraction part
    ax2.add_patch(Rectangle((0, 0), frac2_num/frac2_den, 1, 
                          color='red', alpha=0.3))
    
    # Add grid lines for denominator
    for i in range(1, frac2_den):
        plt.axvline(x=i/frac2_den, color='gray', linestyle='--', alpha=0.3)
    
    plt.title(f'{frac2_num}/{frac2_den}')
    plt.xlim(-0.1, 1.1)
    plt.ylim(-0.1, 1.1)
    
    # 3. Multiplication result
    plt.subplot(133)
    ax3 = plt.gca()
    # Draw full square
    ax3.add_patch(Rectangle((0, 0), 1, 1, fill=False, color='black'))
    
    # Draw both fractions for overlap
    ax3.add_patch(Rectangle((0, 0), 1, frac1_num/frac1_den, 
                          color='blue', alpha=0.3))
    ax3.add_patch(Rectangle((0, 0), frac2_num/frac2_den, 1, 
                          color='red', alpha=0.3))
    
    # Draw overlap area
    ax3.add_patch(Rectangle((0, 0), frac2_num/frac2_den, frac1_num/frac1_den, 
                          color='purple', alpha=0.3))
    
    # Add grid lines
    for i in range(1, max(frac1_den, frac2_den)):
        plt.axhline(y=i/frac1_den, color='gray', linestyle='--', alpha=0.3)
        plt.axvline(x=i/frac2_den, color='gray', linestyle='--', alpha=0.3)
    
    result = (frac1_num/frac1_den) * (frac2_num/frac2_den)
    plt.title(f'Result: {result:.3f}\n({frac1_num}/{frac1_den}  {frac2_num}/{frac2_den})')
    plt.xlim(-0.1, 1.1)
    plt.ylim(-0.1, 1.1)
    
    plt.tight_layout()
    plt.show()

# Example visualizations
visualize_fraction_multiplication(1, 5, 1, 2)  # 1/5  1/2
visualize_fraction_multiplication(2, 3, 3, 4)  # 2/3  3/4
visualize_fraction_multiplication(1, 4, 1, 3)  # 1/4  1/3
```

```python
import numpy as np
import matplotlib.pyplot as plt

#   
x = np.linspace(0.1, 4, 200)  # 0  ( )

def f1(x):
    return x**2

def f2(x):
    return x

#  
quotient = f1(x) / f2(x)  # = x

plt.figure(figsize=(15, 10))

# 1.  
plt.subplot(221)
plt.plot(x, f1(x), 'b-', label='f(x) = x')
plt.plot(x, f2(x), 'r-', label='f(x) = x')
plt.grid(True)
plt.legend()
plt.title('Original Functions')

# 2.  
plt.subplot(222)
plt.plot(x, quotient, 'g-', label='f(x)/f(x) = x')
plt.grid(True)
plt.legend()
plt.title('Division Result')

# 3.     
plt.subplot(223)
x_point = 2
y1 = f1(x_point)
y2 = f2(x_point)
ratio = y1/y2

plt.plot(x, f1(x), 'b-', alpha=0.3, label='f(x)')
plt.plot(x, f2(x), 'r-', alpha=0.3, label='f(x)')
plt.vlines(x_point, 0, y1, 'blue', linestyles='--')
plt.vlines(x_point, 0, y2, 'red', linestyles='--')
plt.plot([x_point], [y1], 'bo')
plt.plot([x_point], [y2], 'ro')

#   
plt.plot([x_point-0.5, x_point+0.5], [y1, y1], 'b-', linewidth=3)
plt.plot([x_point-0.5, x_point+0.5], [y2, y2], 'r-', linewidth=3)
plt.text(x_point+0.6, (y1+y2)/2, f'Ratio = {ratio:.2f}')

plt.grid(True)
plt.legend()
plt.title(f'Division at x={x_point}')

# 4.    
plt.subplot(224)
x_points = [1, 1.5, 2, 2.5, 3]
for x_p in x_points:
    y1 = f1(x_p)
    y2 = f2(x_p)
    ratio = y1/y2
    
    plt.vlines(x_p, y2, y1, 'gray', linestyles='--', alpha=0.5)
    plt.plot([x_p-0.1, x_p+0.1], [y1, y1], 'b-', linewidth=2)
    plt.plot([x_p-0.1, x_p+0.1], [y2, y2], 'r-', linewidth=2)
    plt.text(x_p+0.15, (y1+y2)/2, f'{ratio:.1f}')

plt.plot(x, f1(x), 'b-', alpha=0.3, label='f(x)')
plt.plot(x, f2(x), 'r-', alpha=0.3, label='f(x)')
plt.plot(x, quotient, 'g-', label='f(x)/f(x)')

plt.grid(True)
plt.legend()
plt.title('Ratios at Multiple Points')

plt.tight_layout()
plt.show()

#   :    ()
plt.figure(figsize=(15, 5))

x = np.linspace(0.1, 2*np.pi-0.1, 200)  #  
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

plt.subplot(131)
plt.plot(x, sin_x, 'b-', label='sin(x)')
plt.plot(x, cos_x, 'r-', label='cos(x)')
plt.grid(True)
plt.legend()
plt.title('sin(x) and cos(x)')

plt.subplot(132)
plt.plot(x, tan_x, 'g-', label='tan(x) = sin(x)/cos(x)')
plt.grid(True)
plt.legend()
plt.title('Division Result: tan(x)')

plt.subplot(133)
x_points = [np.pi/6, np.pi/4, np.pi/3]
for x_p in x_points:
    y1 = np.sin(x_p)
    y2 = np.cos(x_p)
    ratio = np.tan(x_p)
    
    plt.vlines(x_p, min(y1,y2), max(y1,y2), 'gray', linestyles='--', alpha=0.5)
    plt.plot([x_p-0.1, x_p+0.1], [y1, y1], 'b-', linewidth=2)
    plt.plot([x_p-0.1, x_p+0.1], [y2, y2], 'r-', linewidth=2)
    plt.text(x_p+0.15, (y1+y2)/2, f'{ratio:.2f}')

plt.plot(x, sin_x, 'b-', alpha=0.3, label='sin(x)')
plt.plot(x, cos_x, 'r-', alpha=0.3, label='cos(x)')
plt.grid(True)
plt.legend()
plt.title('Ratio Visualization')

plt.tight_layout()
plt.show()
```
