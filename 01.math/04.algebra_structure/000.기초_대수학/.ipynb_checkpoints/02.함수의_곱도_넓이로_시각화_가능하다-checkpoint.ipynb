```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# 1. 먼저 숫자의 곱을 1차원과 2차원에서 비교
plt.figure(figsize=(15, 5))

# 1차원에서 2*3=6
plt.subplot(121)
plt.plot([0, 2], [0, 0], 'b-', linewidth=3, label='2')
plt.plot([0, 3], [0.2, 0.2], 'r-', linewidth=3, label='3')
plt.plot([0, 6], [0.4, 0.4], 'g-', linewidth=3, label='2*3=6')
plt.grid(True)
plt.legend()
plt.title('1D Representation of 2*3=6')
plt.ylim(-0.2, 0.6)

# 2차원에서 2*3=6 (면적)
plt.subplot(122)
rect = plt.Rectangle((0, 0), 2, 3, facecolor='green', alpha=0.3)
plt.gca().add_patch(rect)
plt.plot([0, 2], [0, 0], 'b-', linewidth=2, label='2')
plt.plot([0, 0], [0, 3], 'r-', linewidth=2, label='3')
plt.grid(True)
plt.legend()
plt.title('2D Representation of 2*3=6 (Area)')
plt.axis('equal')
plt.xlim(-0.5, 3)
plt.ylim(-0.5, 4)

plt.show()

# 2. 함수의 곱을 3D로 시각화
x = np.linspace(-2, 2, 100)
def f1(x): return 0.5*x + 1  # y = 0.5x + 1
def f2(x): return x + 0.5    # y = x + 0.5

# 2D에서의 표현
plt.figure(figsize=(15, 5))

plt.subplot(121)
plt.plot(x, f1(x), 'b-', label='f₁(x) = 0.5x + 1')
plt.plot(x, f2(x), 'r-', label='f₂(x) = x + 0.5')
plt.plot(x, f1(x)*f2(x), 'g-', label='f₁(x)×f₂(x)')
plt.grid(True)
plt.legend()
plt.title('2D: Product as Single Function')

# 3D에서의 표현
ax = plt.subplot(122, projection='3d')

# 함수들 그리기
ax.plot(x, f1(x), np.zeros_like(x), 'b-', label='f₁(x)')
ax.plot(x, np.zeros_like(x), f2(x), 'r-', label='f₂(x)')

# 각 x에서 직사각형 그리기
for i in range(0, len(x), 10):  # 10개 간격으로 그리기
    xi = x[i]
    y1 = f1(xi)
    y2 = f2(xi)
    
    # 직사각형의 각 모서리
    vertices = [
        [xi, 0, 0],
        [xi, y1, 0],
        [xi, y1, y2],
        [xi, 0, y2]
    ]
    
    # 직사각형 면 그리기
    poly = Poly3DCollection([vertices], alpha=0.1)
    poly.set_color('green')
    ax.add_collection3d(poly)
    
    # 직사각형 테두리 그리기
    ax.plot([xi, xi], [0, y1], [0, 0], 'b-', alpha=0.2)
    ax.plot([xi, xi], [0, 0], [0, y2], 'r-', alpha=0.2)
    ax.plot([xi, xi], [y1, y1], [0, y2], 'g-', alpha=0.2)
    ax.plot([xi, xi], [0, y1], [y2, y2], 'g-', alpha=0.2)

ax.set_xlabel('x')
ax.set_ylabel('f₁(x)')
ax.set_zlabel('f₂(x)')
ax.view_init(elev=20, azim=45)
plt.title('3D: Product as Area')

plt.tight_layout()
plt.show()

# 3. 특정 지점에서의 곱을 자세히 보기
plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')

x_point = 1.0
y1 = f1(x_point)
y2 = f2(x_point)

# 직사각형 테두리 그리기
ax.plot([x_point, x_point], [0, y1], [0, 0], 'b-', linewidth=2)
ax.plot([x_point, x_point], [0, 0], [0, y2], 'r-', linewidth=2)
ax.plot([x_point, x_point], [y1, y1], [0, y2], 'g-', linewidth=2)
ax.plot([x_point, x_point], [0, y1], [y2, y2], 'g-', linewidth=2)

# 직사각형 면 그리기
vertices = np.array([[x_point, 0, 0],
                    [x_point, y1, 0],
                    [x_point, y1, y2],
                    [x_point, 0, y2]])
poly = Poly3DCollection([vertices], alpha=0.3)
poly.set_color('green')
ax.add_collection3d(poly)

ax.set_xlabel('x')
ax.set_ylabel('f₁(x)')
ax.set_zlabel('f₂(x)')
ax.view_init(elev=20, azim=45)
plt.title(f'Product as Area at x={x_point}')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 숫자의 곱과 면적
plt.figure(figsize=(15, 5))

# 숫자의 곱을 면적으로 표현
plt.subplot(131)
a, b = 3, 2
rectangle = plt.Rectangle((0, 0), a, b, facecolor='blue', alpha=0.3)
plt.gca().add_patch(rectangle)
plt.plot([0, a], [0, 0], 'k-', linewidth=2)
plt.plot([0, 0], [0, b], 'k-', linewidth=2)
plt.text(a/2, -0.3, f'{a}', ha='center')
plt.text(-0.3, b/2, f'{b}', va='center')
plt.text(a/2, b/2, f'Area = {a*b}', ha='center')
plt.grid(True)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.title('Numbers: 3 × 2 = 6 (Area)')

# 2. 함수의 곱과 새로운 함수
x = np.linspace(0, 4, 200)

def f1(x):
    return np.sin(x) + 1.5  # Always positive for visualization

def f2(x):
    return np.cos(x/2) + 1.5  # Always positive for visualization

product = f1(x) * f2(x)

plt.subplot(132)
plt.plot(x, f1(x), 'b-', label='f₁(x)')
plt.plot(x, f2(x), 'r-', label='f₂(x)')
plt.plot(x, product, 'g-', label='f₁(x)×f₂(x)')
plt.fill_between(x, 0, product, alpha=0.3, color='green')
plt.grid(True)
plt.legend()
plt.title('Functions: (f₁×f₂)(x)')

# 3. 특정 지점에서의 곱을 면적으로 표현
plt.subplot(133)
x_point = 2
y1 = f1(x_point)
y2 = f2(x_point)
prod = y1 * y2

plt.plot(x, f1(x), 'b-', alpha=0.3, label='f₁(x)')
plt.plot(x, f2(x), 'r-', alpha=0.3, label='f₂(x)')
plt.plot(x, product, 'g-', alpha=0.3, label='f₁(x)×f₂(x)')

# 특정 지점에서의 면적 표시
rectangle = plt.Rectangle((x_point-0.5, 0), 1, prod, facecolor='green', alpha=0.3)
plt.gca().add_patch(rectangle)
plt.vlines(x_point, 0, y1, 'blue', linestyles='--')
plt.vlines(x_point, 0, y2, 'red', linestyles='--')
plt.plot([x_point], [prod], 'ko')

plt.grid(True)
plt.legend()
plt.title('Area at x=2')

plt.tight_layout()
plt.show()

# 적분으로의 확장
plt.figure(figsize=(10, 5))

# 함수의 곱과 적분
plt.plot(x, f1(x), 'b-', label='f₁(x)')
plt.plot(x, f2(x), 'r-', label='f₂(x)')
plt.plot(x, product, 'g-', label='f₁(x)×f₂(x)')

# 구간 [1,3]에서의 적분 영역
x_integral = x[(x>=1) & (x<=3)]
y_integral = product[(x>=1) & (x<=3)]
plt.fill_between(x_integral, 0, y_integral, alpha=0.3, color='green')

integral_value = np.trapz(y_integral, x_integral)
plt.text(2, 6, f'∫f₁f₂dx ≈ {integral_value:.2f}', ha='center')

plt.grid(True)
plt.legend()
plt.title('Integral of Product Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 두 간단한 함수 정의
x = np.linspace(0, 4, 200)

def f1(x):
    return np.sin(x) + 1.5  # 항상 양수

def f2(x):
    return np.cos(x/2) + 1.5  # 항상 양수

product = f1(x) * f2(x)

# 1. 두 함수의 곱을 넓이로 표현
plt.figure(figsize=(15, 5))

# 특정 지점 x=2에서의 곱
x_point = 2
y1 = f1(x_point)
y2 = f2(x_point)
prod = y1 * y2

# 왼쪽: 2D 평면에서의 넓이 표현
plt.subplot(131)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.vlines(x_point, 0, y1, 'blue', linestyles='--', label='f₁(2)')
plt.vlines(x_point, 0, y2, 'red', linestyles='--', label='f₂(2)')
rectangle = plt.Rectangle((x_point-0.5, 0), 1, prod, facecolor='green', alpha=0.3)
plt.gca().add_patch(rectangle)
plt.plot([x_point], [prod], 'ko')
plt.text(x_point+0.1, y1/2, f'{y1:.2f}', color='blue')
plt.text(x_point+0.3, y2/2, f'{y2:.2f}', color='red')
plt.text(x_point, prod+0.2, f'Area = {prod:.2f}', ha='center')
plt.grid(True)
plt.title('Area Representation at x=2')

# 중간: 전체 함수들
plt.subplot(132)
plt.plot(x, f1(x), 'b-', label='f₁(x)')
plt.plot(x, f2(x), 'r-', label='f₂(x)')
plt.plot(x, product, 'g-', label='f₁(x)×f₂(x)')
plt.fill_between(x, 0, product, alpha=0.3, color='green')
plt.plot([x_point], [prod], 'ko')
plt.grid(True)
plt.legend()
plt.title('Full Functions with Product')

# 오른쪽: 순수 수식 표현
plt.subplot(133)
plt.plot(x, product, 'g-', label='f₁(x)×f₂(x)')
plt.plot([x_point], [prod], 'ko')
plt.text(x_point+0.1, prod+0.2, f'{prod:.2f}')
plt.grid(True)
plt.legend()
plt.title('Pure Function Value')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values
x = np.linspace(-3, 3, 200)

# Define two simple functions
def f1(x):
    return np.sin(x)

def f2(x):
    return np.exp(-x**2/2)

# Calculate product
product = f1(x) * f2(x)

# Create plots
plt.figure(figsize=(15, 5))

# Plot both functions
plt.subplot(131)
plt.plot(x, f1(x), label='f₁(x) = sin(x)')
plt.plot(x, f2(x), label='f₂(x) = e^(-x²/2)')
plt.title('Original Functions')
plt.grid(True)
plt.legend()

# Plot product
plt.subplot(132)
plt.plot(x, product, 'r', label='f₁(x) × f₂(x)')
plt.fill_between(x, 0, product, alpha=0.3)
plt.title('Product (with area)')
plt.grid(True)
plt.legend()

# Show vertical multiplication
plt.subplot(133)
plt.plot(x, f1(x), label='f₁(x)', alpha=0.5)
plt.plot(x, f2(x), label='f₂(x)', alpha=0.5)
plt.plot(x, product, 'r', label='Product')

# Add vertical lines at specific points to show multiplication
x_points = [-2, -1, 0, 1, 2]
for x_val in x_points:
    y1 = f1(x_val)
    y2 = f2(x_val)
    prod = y1 * y2
    plt.vlines(x_val, 0, prod, colors='gray', linestyles='--', alpha=0.5)
    plt.plot([x_val], [prod], 'ko')

plt.title('Vertical Multiplication')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# First, let's show the polynomial multiplication
x = np.linspace(-2, 2, 200)

def f1(x):
    return x**2

def f2(x):
    return x**3 + 1

product = f1(x) * f2(x)  # This will be x^5 + x^2

plt.figure(figsize=(15, 5))

# Original functions
plt.subplot(131)
plt.plot(x, f1(x), label='f₁(x) = x²')
plt.plot(x, f2(x), label='f₂(x) = x³ + 1')
plt.title('Original Functions')
plt.grid(True)
plt.legend()

# Product
plt.subplot(132)
plt.plot(x, product, 'r', label='f₁(x) × f₂(x) = x⁵ + x²')
plt.fill_between(x, 0, product, alpha=0.3)
plt.title('Product (with area)')
plt.grid(True)
plt.legend()

# Vertical multiplication
plt.subplot(133)
plt.plot(x, f1(x), label='f₁(x)', alpha=0.5)
plt.plot(x, f2(x), label='f₂(x)', alpha=0.5)
plt.plot(x, product, 'r', label='Product')

# Add vertical lines at specific points
x_points = [-1.5, -0.5, 0, 0.5, 1.5]
for x_val in x_points:
    y1 = f1(x_val)
    y2 = f2(x_val)
    prod = y1 * y2
    plt.vlines(x_val, 0, prod, colors='gray', linestyles='--', alpha=0.5)
    plt.plot([x_val], [prod], 'ko')

plt.title('Vertical Multiplication')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Now let's show 2D function multiplication
# Create grid of x and y values
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Define two 2D functions
def f1_2d(x, y):
    return (x**2 + y**2)  # Paraboloid

def f2_2d(x, y):
    return np.sin(x) * np.cos(y)  # Wave pattern

# Calculate product
Z1 = f1_2d(X, Y)
Z2 = f2_2d(X, Y)
Z_product = Z1 * Z2

# Create 3D plots
fig = plt.figure(figsize=(15, 5))

# Plot first function
ax1 = fig.add_subplot(131, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z1, cmap='viridis')
ax1.set_title('f₁(x,y) = x² + y²')
plt.colorbar(surf1, ax=ax1)

# Plot second function
ax2 = fig.add_subplot(132, projection='3d')
surf2 = ax2.plot_surface(X, Y, Z2, cmap='viridis')
ax2.set_title('f₂(x,y) = sin(x)cos(y)')
plt.colorbar(surf2, ax=ax2)

# Plot product
ax3 = fig.add_subplot(133, projection='3d')
surf3 = ax3.plot_surface(X, Y, Z_product, cmap='viridis')
ax3.set_title('f₁(x,y) × f₂(x,y)')
plt.colorbar(surf3, ax=ax3)

plt.tight_layout()
plt.show()

# Show contour plots for better understanding
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.contour(X, Y, Z1, levels=20)
plt.colorbar(label='Value')
plt.title('f₁(x,y) contours')
plt.grid(True)

plt.subplot(132)
plt.contour(X, Y, Z2, levels=20)
plt.colorbar(label='Value')
plt.title('f₂(x,y) contours')
plt.grid(True)

plt.subplot(133)
plt.contour(X, Y, Z_product, levels=20)
plt.colorbar(label='Value')
plt.title('Product contours')
plt.grid(True)

plt.tight_layout()
plt.show()
```
