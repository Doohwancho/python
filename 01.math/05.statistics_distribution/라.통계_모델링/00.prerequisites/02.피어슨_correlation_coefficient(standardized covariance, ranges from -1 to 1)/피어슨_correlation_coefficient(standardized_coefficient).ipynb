```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_correlations():
    # Create different correlations
    np.random.seed(42)
    n_points = 100
    
    # Generate x values
    x = np.random.normal(0, 1, n_points)
    
    # Generate y values with different correlations
    y_perfect_pos = x  # correlation = 1
    y_strong_pos = x + np.random.normal(0, 0.5, n_points)  # correlation ≈ 0.8
    y_moderate_pos = x + np.random.normal(0, 1, n_points)  # correlation ≈ 0.5
    y_no_corr = np.random.normal(0, 1, n_points)  # correlation ≈ 0
    y_perfect_neg = -x  # correlation = -1
    
    # Create subplots
    fig, axes = plt.subplots(1, 5, figsize=(20, 4))
    
    # Plot each correlation
    data = [
        (y_perfect_pos, "Perfect Positive\nr = 1.0"),
        (y_strong_pos, "Strong Positive\nr ≈ 0.8"),
        (y_moderate_pos, "Moderate Positive\nr ≈ 0.5"),
        (y_no_corr, "No Correlation\nr ≈ 0"),
        (y_perfect_neg, "Perfect Negative\nr = -1.0")
    ]
    
    for ax, (y, title) in zip(axes, data):
        # Plot points
        ax.scatter(x, y, alpha=0.5)
        
        # Add correlation line
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        ax.plot(x, p(x), "r--", alpha=0.8)
        
        # Calculate actual correlation
        corr = np.corrcoef(x, y)[0,1]
        
        ax.set_title(f"{title}\nActual r = {corr:.2f}")
        ax.grid(True, alpha=0.3)
        ax.set_xlabel("X")
        ax.set_ylabel("Y")
    
    plt.tight_layout()
    plt.show()

visualize_correlations()
```
