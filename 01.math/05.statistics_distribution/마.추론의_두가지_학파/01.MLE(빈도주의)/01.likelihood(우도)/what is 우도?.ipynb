```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# 데이터 설정
n_trials = 10  # 총 시행 횟수
n_heads = 7    # 관찰된 앞면 횟수

# 확률값 범위 생성
theta = np.linspace(0, 1, 1000)

# 우도 함수 계산
likelihood = binom.pmf(n_heads, n_trials, theta)

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(theta, likelihood, 'b-', lw=2, label='Likelihood')
plt.fill_between(theta, likelihood, alpha=0.2)
```

```python
# 최대 우도 추정치 표시
mle = n_heads / n_trials
plt.axvline(mle, color='r', linestyle='--', label=f'MLE (θ = {mle:.2f})')

plt.title('이항 분포의 우도 함수')
plt.xlabel('θ (앞면이 나올 확률)')
plt.ylabel('Likelihood')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 여러 데이터에 대한 우도 비교
plt.figure(figsize=(12, 6))
for heads, trials in [(7, 10), (14, 20), (70, 100)]:
    likelihood = binom.pmf(heads, trials, theta)
    plt.plot(theta, likelihood, label=f'{trials}번 중 {heads}번 앞면')

plt.title('다양한 데이터에 대한 우도 함수 비교')
plt.xlabel('θ (앞면이 나올 확률)')
plt.ylabel('Likelihood')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 로그 우도 계산
log_likelihood = binom.logpmf(n_heads, n_trials, theta)

plt.figure(figsize=(10, 6))
plt.plot(theta, log_likelihood, 'g-', lw=2, label='Log-Likelihood')
plt.axvline(mle, color='r', linestyle='--', label=f'MLE (θ = {mle:.2f})')

plt.title('이항 분포의 로그 우도 함수')
plt.xlabel('θ (앞면이 나올 확률)')
plt.ylabel('Log-Likelihood')
plt.legend()
plt.grid(True)
plt.show()
```
